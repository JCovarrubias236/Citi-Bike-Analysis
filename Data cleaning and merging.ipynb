```python
import pandas as pd
```

```python
# Upload data into their own dataframes
df1 = pd.read_csv('Raw Data/JC-202301-citibike-tripdata.csv')

df2 = pd.read_csv('Raw Data/JC-202302-citibike-tripdata.csv')

df3 = pd.read_csv('Raw Data/JC-202303-citibike-tripdata.csv')

df4 = pd.read_csv('Raw Data/JC-202304-citibike-tripdata.csv')

df5 = pd.read_csv('Raw Data/JC-202305-citibike-tripdata.csv')

df6 = pd.read_csv('Raw Data/JC-202306-citibike-tripdata.csv')

df7 = pd.read_csv('Raw Data/JC-202307-citibike-tripdata.csv')

df8 = pd.read_csv('Raw Data/JC-202308-citibike-tripdata.csv')

df9 = pd.read_csv('Raw Data/JC-202309-citibike-tripdata.csv')

df10 = pd.read_csv('Raw Data/JC-202310-citibike-tripdata.csv')

df11 = pd.read_csv('Raw Data/JC-202311-citibike-tripdata.csv')

df12 = pd.read_csv('Raw Data/JC-202312-citibike-tripdata.csv')

df13 = pd.read_csv('Raw Data/JC-202401-citibike-tripdata.csv')

df14 = pd.read_csv('Raw Data/JC-202402-citibike-tripdata.csv')

df15 = pd.read_csv('Raw Data/JC-202403-citibike-tripdata.csv')
```

```python
full_data = [df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15]

results = pd.concat(full_data)

results.info()
```

```python
# Remove duplicates
results = results.dropna()
results.info()
```

```python
results.head(10)
```

```python
# Convert 'started_at' and 'ended_at' to datetime columns
results[['started_at', 'ended_at']] = results[['started_at', 'ended_at']].apply(pd.to_datetime)
results.info()
```

```python
results.to_csv('Clean and merged dataset/1-1-2023_to_3-31-2024_citibike_tripdata.csv')
```
